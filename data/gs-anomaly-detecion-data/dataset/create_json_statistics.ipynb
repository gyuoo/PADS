### 초기 모평균 모표준편차 생성을 위한 노트북

```python
import numpy as np
import pandas as pd
import json
```

```python
data_route = "C:\\workspace\\project_final\\data\\dataset_sample_10percent.csv"
data = pd.read_csv(data_route, encoding='utf-8')
```

```python
data['brandclass_name'] = data['brand_name'] + '_' + data['class_name']

# 그룹화 기준 컬럼
group_columns = ['supplier_code', 'cate2_nm', 'brandclass_name']
```

```python
# 각 그룹별 평균과 표준편차 계산 후 별도 파일로 저장
for col in group_columns:
    # 그룹별 평균과 표준편차 계산
    group_stats = data.groupby(col)['price'].agg(['mean', 'std']).reset_index()
    group_stats['count'] = data[col].value_counts().reindex(group_stats[col]).fillna(0).astype(int).values
    
    # 결과를 리스트 형식으로 저장
    result = []
    for _, row in group_stats.iterrows():
        result.append({
            "name": row[col],
            "count": int(row['count']),
            "mean": row['mean'],
            "std": 0 if pd.isna(row['std']) else row['std']
        })
        
    # JSON 파일로 저장
    output_file = f'C:\\workspace\\project_final\\data\\{col}_statistics.json'
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(result, f, ensure_ascii=False, indent=4)

    # 저장된 JSON 출력 확인
    print(f"File saved: {output_file}")
    with open(output_file, 'r', encoding='utf-8') as f:
        saved_data = json.load(f)
    print(json.dumps(saved_data, ensure_ascii=False, indent=4))
```
